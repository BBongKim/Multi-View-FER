```python
import os
import shutil
```

```python
listdir.remove('auto-classifier.ipynb')
listdir.remove('.ipynb_checkpoints')
listdir.remove('.DS_Store')
```

```python
def createFolder(directory):
    print(directory)
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory. ' + directory)
```

```python
def classifyFileWithAngle(fileDir, fileName):
    print(fileDir)
    if len(fileName) == 7:
        emotion = fileName[4:6]
    elif len(fileName) == 8:
        emotion = fileName[4:6]

    if emotion == "AF":
        shutil.move(fileDir, currentDir + "/afraid")
    elif emotion == "AN":
        shutil.move(fileDir, currentDir + "/angry")
    elif emotion == "DI":
        shutil.move(fileDir, currentDir + "/disgusted")
    elif emotion == "HA":
        shutil.move(fileDir, currentDir + "/happy")
    elif emotion == "NE":
        shutil.move(fileDir, currentDir + "/neutral")
    elif emotion == "SA":
        shutil.move(fileDir, currentDir + "/sad")
    elif emotion == "SU":
        shutil.move(fileDir, currentDir + "/surprised")
```

```python
def classifyFile(fileDir, fileName):
    print(fileDir)
    emotion = fileName[4:6]
    print(emotion)

    if emotion == "AF":
        afraidDir = currentDir + "/afraid"
        createFolder(afraidDir)
        shutil.move(fileDir, afraidDir)

    elif emotion == "AN":
        angryDir = currentDir + "/angry"
        createFolder(angryDir)
        shutil.move(fileDir, angryDir)

    elif emotion == "DI":
        disgustedDir = currentDir + "/disgusted"
        createFolder(disgustedDir)
        shutil.move(fileDir, disgustedDir)

    elif emotion == "HA":
        happyDir = currentDir + "/happy"
        createFolder(happyDir)
        shutil.move(fileDir, happyDir)

    elif emotion == "NE":
        neutralDir = currentDir + "/neutral"
        createFolder(neutralDir)
        shutil.move(fileDir, neutralDir)

    elif emotion == "SA":
        sadDir = currentDir + "/sad"
        createFolder(sadDir)
        shutil.move(fileDir, sadDir)

    elif emotion == "SU":
        surprisedDir = currentDir + "/surprised"
        createFolder(surprisedDir)
        shutil.move(fileDir, surprisedDir)
```

```python
for folderName in listdir:
    imagefolderDir = currentDir + "/" + folderName
    files = os.listdir(imagefolderDir)
    for fileName in files:
        classifyFile(imagefolderDir + "/" + fileName, fileName)
```
